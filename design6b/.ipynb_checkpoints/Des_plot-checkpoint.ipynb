```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import spline
%matplotlib inline
import csv
import pandas as pd
import matplotlib
```

```python
y1 = np.genfromtxt('dmd11.txt',delimiter=' ')*1000
y2 = np.genfromtxt('dmd21.txt',delimiter=' ')*1000
```

```python
x1=np.arange(1,len(y1)+1,1)
x2=np.arange(1,len(y1)+1,1)
```

```python
xa=np.linspace(x1.min(),x1.max(),1000)
xb=np.linspace(x2.min(),x2.max(),1000)
ya=spline(x1,y1,xa)
yb=spline(x2,y2,xb)
plt.plot(xa,ya)
plt.plot(xb,yb)
plt.xlabel("k")
plt.ylabel("DMD [ps\km]")
plt.title("DMD VS K ")
plt.grid()
plt.show()
```

```python
plt.figure()
#plt.rcParams.update({'font.size': 8})
x1_smooth=np.linspace(x1.min(),x1.max(),300)
y1_smooth=spline(x1,y1,x1_smooth)
plt.subplot(1,2,1)
plt.plot(x1_smooth,y1_smooth)
plt.xlabel("Alpha")
plt.ylabel("DMD [ps\km]")
plt.title("DMD VS Alpha ")

x2_smooth=np.linspace(x2.min(),x2.max(),300)
y2_smooth=spline(x2,y2,x2_smooth)
plt.subplot(1,2,2)
plt.plot(x2_smooth,y2_smooth)
plt.xlabel("Alpha")
plt.ylabel("DMD [ps\km]")
plt.title("DMD VS Alpha ")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.rcParams.update({'font.size': 12})
y_formatter = matplotlib.ticker.ScalarFormatter(useOffset=False)

neff0 = np.genfromtxt('neff_01.txt',delimiter=' ')
neff1 = np.genfromtxt('neff_11.txt',delimiter=' ')
neff2 = np.genfromtxt('neff_21.txt',delimiter=' ')
#splie fit
neff0_smooth=spline(x1,neff0,x1_smooth)
neff1_smooth=spline(x1,neff1,x1_smooth)
neff2_smooth=spline(x1,neff2,x1_smooth)

plt.plot(x1_smooth,neff0_smooth)
plt.plot(x1_smooth,neff1_smooth)
plt.plot(x1_smooth,neff2_smooth)
plt.xlabel("k")
plt.ylabel("DMD [ns\km]")
plt.title("DMD VS K ")
plt.gca().yaxis.set_major_formatter(y_formatter)
#plt.show()
###########

plt.xlabel("Alpha")
plt.ylabel("N Effective")
plt.title("N Effective Vs Aplha ")
plt.gca().yaxis.set_major_formatter(y_formatter)
plt.tight_layout()
plt.show()
```
