```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import spline
%matplotlib inline
import csv
import pandas as pd
import matplotlib
```

```python
y1pos = np.genfromtxt('dmd11.txt',delimiter=' ')
y2pos = np.genfromtxt('dmd21.txt',delimiter=' ')
#y1neg = np.genfromtxt('dmd11neg.txt',delimiter=' ')
#y2neg = np.genfromtxt('dmd21neg.txt',delimiter=' ')
```

```python
x1pos=np.arange(3,9.5,.5)
#x1neg=np.arange(-6.2,0,.2)
x2pos=np.arange(3,9.5,.5)
#x2neg=np.arange(-6.2,0,.2)
```

```python
y1=y1pos #np.append(y1neg,y1pos)
x1=x1pos #np.append(x1neg,x1pos)
y2=y2pos #np.append(y2neg,y2pos)
x2=x2pos #np.append(x2neg,x2pos)
```

```python
xa=np.linspace(x1.min(),x1.max(),1000)
xb=np.linspace(x2.min(),x2.max(),1000)
ya=spline(x1,y1,xa)
yb=spline(x2,y2,xb)
plt.plot(xa,ya)
plt.plot(xb,yb)
plt.xlabel("k")
plt.ylabel("DMD [ns\km]")
plt.title("DMD VS K ")
plt.grid()
plt.show()
```

```python
plt.figure()
plt.rcParams.update({'font.size': 8})

x1pos_smooth=np.linspace(x1pos.min(),x1pos.max(),300)
y1pos_smooth=spline(x1pos,y1pos,x1pos_smooth)
plt.subplot(2,2,1)
plt.plot(x1pos_smooth,y1pos_smooth)
plt.xlabel("k")
plt.ylabel("DMD [ns\km]")
plt.title("DMD VS K ")

x2pos_smooth=np.linspace(x2pos.min(),x2pos.max(),300)
y2pos_smooth=spline(x2pos,y2pos,x2pos_smooth)
plt.subplot(2,2,2)
plt.plot(x2pos_smooth,y2pos_smooth)
plt.xlabel("k")
plt.ylabel("DMD [ns\km]")
plt.title("DMD VS K ")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
y_formatter = matplotlib.ticker.ScalarFormatter(useOffset=False)

neff0 = np.genfromtxt('neff_01.txt',delimiter=' ')
neff1 = np.genfromtxt('neff_11.txt',delimiter=' ')
neff2 = np.genfromtxt('neff_21.txt',delimiter=' ')
#splie fit
neff0_smooth=spline(x1pos,neff0,x1pos_smooth)
neff1_smooth=spline(x1pos,neff1,x1pos_smooth)
neff2_smooth=spline(x1pos,neff2,x1pos_smooth)

plt.subplot(1,2,1)
plt.plot(x1pos_smooth,neff0_smooth)
plt.plot(x1pos_smooth,neff1_smooth)
plt.plot(x1pos_smooth,neff2_smooth)
plt.xlabel("k")
plt.ylabel("DMD [ns\km]")
plt.title("DMD VS K ")
plt.gca().yaxis.set_major_formatter(y_formatter)
#plt.show()
###########

plt.tight_layout()
plt.show()
```
